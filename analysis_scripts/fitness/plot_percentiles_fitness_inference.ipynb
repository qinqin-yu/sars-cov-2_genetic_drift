```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
font = {'family' : 'arial',
        'size'   : 15}
matplotlib.rc('font', **font)
import seaborn as sns
sns.set(style = 'whitegrid', font_scale = 1.2)
```

```python
conservative = pd.read_csv('../../data/fitness/concat_fitness.csv', index_col = 0)
# hmm = pd.read_csv('../../data/fitness/concat_fitness_HMM.csv', index_col = 0)
```

```python
plt.hist(np.abs(conservative['s']), 1000, cumulative = True, histtype = 'step', density = True, label = 'Conservative, determinstic fitness')
# plt.hist(np.abs(hmm['s']), 10000, cumulative = True, histtype = 'step', density = True, label = 'HMM fitness')
plt.xlabel('Absolute value of fitness effect, |s|')
plt.ylabel('CDF')
plt.legend()
plt.xlim([0,0.5])
plt.tight_layout()
plt.savefig('../../figures/figure_outputs/inferred_fitness_cdf.png')
plt.savefig('../../figures/figure_outputs/inferred_fitness_cdf.pdf')
plt.show()
```

```python
print('Conservative, deterministic fitness')
print('50th percentile', np.quantile(np.abs(conservative['s']), 0.5))
print('75th percentile', np.quantile(np.abs(conservative['s']), 0.75))
print('90th percentile', np.quantile(np.abs(conservative['s']), 0.90))

# print('\nHMM fitness')
# print('50th percentile', np.quantile(np.abs(hmm['s']), 0.5))
# print('75th percentile', np.quantile(np.abs(hmm['s']), 0.75))
# print('90th percentile', np.quantile(np.abs(hmm['s']), 0.90))
```
